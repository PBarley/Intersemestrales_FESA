##### 1. Si la edad es null (na) llenar el campo con la medida de la edad (usar el método fillna())

```python
df_ti['age'].describe()
```

```python
df_ti['age']= df_ti.age.fillna(df_ti.age.mean())
df_ti
```

##### 2. Si embarqued es na llenar con s.

```python
df_ti['embarked']= df_ti.embarked.fillna('S')
df_ti
```

3. Cambiar los valores alfanuméricos de sex a número usando {'female' : 1, 'male' : 0}

```python
df_ti['sex'] = df_ti['sex'].replace({'female' : 1, 'male' : 0})
df_ti
```

4. Cambiar los valores alfanuméricos de embarked a número usando {'S' : 0, 'C':1, 'Q':2}

```python
df_ti['embarked'] = df_ti['embarked'].replace({'S' : 0, 'C':1, 'Q':2})
df_ti
```

5. Crear una nueva variable llamada child con valor 1 para los pasajeros con una edad menor a 16 años y 0 si el pasajero tiene 16 años o mas

```python
df_ti['child'] = (df_ti['age'] < 16).astype(int)
df_ti
```

6. Eliminar las columnas ticket, cabin y name

```python
df_ti = df_ti.drop(['ticket', 'cabin', 'name'], axis=1)
df_ti
```

7. Validar que no existan registros duplicados

```python
df_ti.columns.tolist()
```

```python
df_ti_dup=df_ti.duplicated(subset=[ 'sex', 'age', 'sibsp', 'parch', 'fare','embarked'], keep='last')
df_ti_dup
```
