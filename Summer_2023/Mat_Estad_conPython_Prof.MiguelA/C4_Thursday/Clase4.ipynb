```python
import matplotlib.pyplot as plt
import numpy as np
```

### Generar n√∫meros aleatorios

```python
rnd = np.random.default_rng(4567)
datos = np.arange(15)
print(datos)

probabilidades = np.array(
    [0.3, 0.2, 0.1, 0.05, .05, 0.05, 0.05, 0.025,
    0.025, 0.025, 0.025, 0.025, 0.025, 0.025, 0.025]
)

assert round(sum(probabilidades), 10) == 1.0, "La probabilidad debe sumar 1"
seleccion = rnd.choice(datos, p = probabilidades, replace = True)
print("\n", seleccion)
seleccion = rnd.choice(datos, p = probabilidades, replace = True, size = (4,3))
print("\n", seleccion)
```

```python
rndDos = np.random.default_rng(2456)
rndFlotantes = rndDos.random(size=(3,3))
print(rndFlotantes)

rndF = np.random.uniform(1.2, 1.60, size = 10)
print("\n", rndF)

rndenteros = rndDos.integers(5, 15, size = 8)
print("\n", rndenteros)

distribucion = rndDos.integers(1, 10, size = 20)

fig, ax = plt.subplots()
ax.hist(distribucion)
ax.set_title("Historia de numeros aleatorios")
ax.set_xlabel("Valor")
ax.set_ylabel("Frecuencia")

plt.show()
print(distribucion)
```

```python
rnd = np.random.default_rng(12345)
mu = 5.0 #Valor Medio
sigma = 3 #Desviacion estandar

rnds = rnd.normal(loc = mu, scale = sigma, size = 1000)
fig, ax = plt.subplots()
ax.hist(rnds, bins = 20) # bins = Contenedor = Rango de clases
ax.set_axisbelow(True)
ax.yaxis.grid(color='gray', linestyle='dashed')
ax.hist(rnds,bins=20,linewidth=0.5, edgecolor="white")
plt.show()
```
