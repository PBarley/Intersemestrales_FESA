### 1. Los datos siguientes muestran los pesos de 50 estudiantes que se registraron con aproximación de una libra.
159 136 180 129 162 150 129 146 128 178 144 167 132 172 156 178 129 142
146 159 131 135 168 152 163 145 173 147 137 143 142 133 152 166 146 131
131 160 162 146 151 179 146 142 172 175 166 137 162 159

De los datos de arriba encontrar la siguiente información:

```python
#Declaración de los datosen un arreglo
data1 = np.array([159,136,180,129,162,150,129,146,128,178,144,167,132,172,
                  156,178,129,142,146,159,131,135,168,152,163,145,173,147,
                  137,143,142,133,152,166,146,131,131,160,162,146,151,179,
                  146,142,172,175,166,137,162,159])
data1or = np.sort(data1) #Ordena de menor a mayor
#print(data1, "\n\n", data1or)
print("Sea el conjunto de datos:\n", data1, "\n\n")

#• El valor más grande
print("El valor más grande es: ", data1or[-1:])

#• El valor más pequeño
print("El valor más pequeño es: ", data1or[:1])

#• Obtuvieron 125(datos 1)
suma_125 = np.where(data1==125,1,0).sum()
print("Estudiantes de 125 libras de peso: ", suma_125)

#• Obtuvieron 138(datos 1)
suma_138 = np.where(data1==138,1,0).sum()
print("Estudiantes de 138 libras de peso: ", suma_138)

#• Porcentaje de valores (129<= x >=146, dato1)
sumPerc = np.where((data1>=129) & (data1<=146),1,0).sum()/len(data1)
print("El % de valores 129<= x >=146 es: ", (sumPerc*100),"%")




#• Frecuencias absolutas simples
variables, repetidas = np.unique(data1or, return_counts = True)
frecuencias_absolutas_simple = repetidas
print("\n\n-------------Frecuencias-------------")
print("Por las variables: ", variables, 
      " se obtiene que: \n\nLa Frecuencia absoluta simple es: \n", 
      frecuencias_absolutas_simple)

#• Frecuencias absolutas acumuladas     
frecuencias_absolutas_acumuladas = np.cumsum(frecuencias_absolutas_simple)
print("\nLa Frecuencia absoluta acumulada es: \n", frecuencias_absolutas_acumuladas)

#• Frecuencias relativas simples
frecuencias_relativas_simple = np.round((frecuencias_absolutas_simple/len(data1)) * 100, 2)
print("\nLa Frecuencia relativa simple es: \n", frecuencias_relativas_simple)

#• Frecuencias relativas acumuladas
frecuencias_relativas_acumuladas = np.cumsum(frecuencias_relativas_simple)
print("\nLa Frecuencia relativa acumulada es: \n", frecuencias_relativas_acumuladas)
print("----------------------------------------\n\n")




#• Media aritmética valores agrupados
ma_agrupados = variables * repetidas
print("\n\nMedia aritmética de los valores agrupados: ",
      (np.sum(ma_agrupados)/len(data1)))

#• Media aritmética de valores no agrupados
ma_no_agrupados = np.mean(data1or)
print("Media aritmética de los valores no agrupados: ", 
     (np.sum(data1or)/len(data1)))
print("Media aritmética de los valores no agrupados (usando Numpy): ", 
     ma_no_agrupados)


#• Media Geométrica
'''
media_geo_agrup=np.power(np.product(variables**repetidas),1/len(data1))
print("Media Geométrica por datos agrupados es: ", media_geo_agrup)

media_geo_NoAgrup=np.power(np.product(data1or),1/len(data1))
print("Media Geométrica por datos no agrupados es: ", media_geo_NoAgrup)
'''
print("Media Geométrica usando statistics es: ",sts.geometric_mean(data1or))


#• Media Armónica H
med_Armonica = sts.harmonic_mean(data1or)
print("La Media Armónica H es: ", med_Armonica)

#• Mediana
med = np.median(data1or)
print("La Media es: ", med)

#• Moda
moda = sts.mode(data1or)
print("La moda es: ", moda)

#• Múltiples Modas (si las hay)
mult_modas = sts.multimode(data1or)
if len(mult_modas) <= 1:
    print("No hay multiples modas. Solo esta la moda: ", moda)
else:
    print("Las multiples modas son: ", mult_modas)

    


print("\n\n\n\n-------------Divisiones Proporcionales-------------")
#• Cuartiles
print("Lo Cuartiles son:\n", 
      [round(q, 2) for q in sts.quantiles(data1or, n=4)])

#• Deciles
print("\nLo Deciles son:\n", 
      [round(q, 2) for q in sts.quantiles(data1or, n=10)])

#• Percentiles
print("\nLo Percentiles son:\n", 
      [round(q, 2) for q in sts.quantiles(data1or, n=100)])
print("----------------------------------------")
```

##### Histograma con Matplotlib

```python
# • Realizar el histograma con Matplotlib

data1hst = np.sort(np.reshape(data1,50))
rango = data1or[49] - data1or[0]
intervalos=rango/13
b=np.arange(data1or[0],data1or[49],intervalos,dtype=np.int16)
fig, ax = plt.subplots()

intervalos_clase=ax.hist(data1or, bins=b,range=(data1or[0],data1or[49]), linewidth=0.5, edgecolor="white", color = "#FFC0CB")
plt.xticks(fontsize=8)
ax.set(xlim=(124, 180), xticks=intervalos_clase[1],
       ylim=(0, 9), yticks=np.arange(0, 9, 1,dtype=np.int16))
ax.yaxis.grid(color='gray', linestyle='dashed')
ax.set_axisbelow(True)
ax.grid(True)
```
