#### 1. Comprobar gráficamente si las siguientes ecuaciones tienen solución.


##### Notas de repaso:
- Empleando np.linspace(): Esta función es un comando de la biblioteca NumPy y se utiliza para crear un arreglo numérico llamado "x" que contiene una secuencia de valores igualmente espaciados.Los parámetros utilizados en la función son:

    El primer parámetro: Especifica el valor inicial de la secuencia.

    El segundo parámetro: Especifica el valor final de la secuencia.

    El tercer parámetro: Indica el número de elementos que se generarán en la secuencia.
    


- Uso de del código: ax.set(xlim =(a,b), ylim =(c,d), xticks=np.arange(e,f,g), yticks=np.arange())

    El código utiliza un objeto de ejes (ax) para establecer varias propiedades del gráfico, como los límites de los ejes, las marcas de los ejes y los rangos permitidos.Desglosando los parámetros utilizados en el método set() se tiene:


        - xlim = (a,b) &  ylim = (a,b): 

        Establecen los límites del eje x & y, respectivamente en un determinado rango. Significa que el eje x (o y) se extenderá desde el valor a hasta el valor b.


        - xticks = np.arange(c,d,e) & yticks = np.arange(c,d,e):

        Estos parámetros especifican las marcas o etiquetas a mostrar en el eje x & y. Utiliza la función np.arange() de NumPy para generar un arreglo de valores desde c hasta d (exclusivo) con un incremento de e. 

```python
# PRIMER SISTEMA DE ECUACIONES

def h1(x):
    return (6+(-3*x))/2

def h2(x):
    return (-9 +(-6*x))/4

print("Resultado: El sistema no tiene solución")

fig,ax = plt.subplots()

x = np.linspace(-10,20) #Aumenta el rango
ec1 = plt.plot(x,h1(x))
ec2 = plt.plot(x,h2(x))

ax.set_title("Grafica del Primer Sistema de Ecuaciones")
ax.legend(("Ec.1", "Ec.2"), )
ax.set_xlabel("Eje x")
ax.set_ylabel("Eje y")
ax.set(xlim = (-10, 20), ylim = (-20, 10), xticks=np.arange(-10,20,10), yticks=np.arange(-20,10,10))
plt.grid()
```

#### 2. Encuentra la ecuación de la recta tangente de la parábola.

En el punto (1,4). Dibuja la parábola y muestra un segmento de la recta tangente en el punto
(1,4).

```python
# Dibujo de la Parabola:
def parab(x):
    return (x**2) + 3



#Ecuación de la recta tangente de la parábola: Punto - Pendiente
punto_tang = np.array([1,4])

def pendiente(x):
    return 2*x

m = pendiente(punto_tang[0])

def lin_tang(x):
    return (m * x) - ((m * punto_tang[0]) - punto_tang[1])


x = np.linspace(-10,10) #Aumenta el rango
fig,ax = plt.subplots()
parabola = plt.plot(x,parab(x))
tangente = plt.plot(x,lin_tang(x))






ax.set_title("Grafica Recta Tangente de la Parábola")
ax.legend(("Parabola", "Tangente"))
ax.set_xlabel("Eje x")
ax.set_ylabel("Eje y")
ax.set(xlim = (0, 6), ylim = (0, 40), xticks=np.arange(0,6,1), yticks=np.arange(0,40,5))
plt.grid()

```
